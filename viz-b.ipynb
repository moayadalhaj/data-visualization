```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

Import needed librarries.

Load the data I want to analyse it and show it.

```python
hotels_df =pd.read_csv('hotel_booking.csv')
hotels_df
```

```python
sns.relplot(x='arrival_date_year',y='adults',data=hotels_df)
```

From this plot the data cannot be plotted properly, we need to group it then plot

```python
adults_df=hotels_df.groupby('arrival_date_year',as_index=False).adults.sum()
adults_df
```

```python
sns.barplot(x='arrival_date_year',y='adults',data=adults_df)
```

So, we can obviously number of adults that bokking hotels in 2016 is the maximmum

```python
sns.relplot(x='arrival_date_month',y='adults',data=hotels_df)
```

Also, in this plot the data cannot be plotted properly, we need to group it then plot

```python
adults_month_df=hotels_df.groupby('arrival_date_month',as_index=False).adults.sum()
adults_month_df
```

```python
graph=sns.barplot(x='arrival_date_month',y='adults',data=adults_month_df)
graph.set_xticklabels(labels=adults_month_df['arrival_date_month'],rotation=45)
```

From this bar chart we can figure that the maximmum nuber of adults was on august then in july.
